```python
A=combine.groupby(['Outlet_Identifier']).count()
A=A[A.columns[0:1]]
A.columns=['Outlet_Count']
combine=pd.merge(combine,A,on='Outlet_Identifier')
```

```python
combine
```

```python
#Item count
B=combine.groupby(['Item_Identifier']).count()
B=B[B.columns[0:1]]
B.columns=['Item_Count']
combine=pd.merge(combine,B,on='Item_Identifier')
combine
```

```python
combine.info()
```

```python
#Counting no. of years of establishment
C=combine['Outlet_Establishment_Year'].unique()
C=pd.DataFrame(C,columns=['Outlet_Establishment_Year'])
C['Outlet_Year']=2020
C['Outlet_Year']=C['Outlet_Year']-C['Outlet_Establishment_Year']
combine=pd.merge(combine,C,on='Outlet_Establishment_Year')
combine
```

```python
combine['Item_Fat_Content'].replace({"Low Fat":0,"Regular":1},inplace=True)
combine
```

```python
#Getting dummy values for Outlet Size
dummies_outlet_size=pd.get_dummies(data=combine,columns=['Outlet_Size','Outlet_Location_Type','Outlet_Type'])
dummies_outlet_size
```

```python
dummies_outlet_size.info()
```

```python
#dropping less important columns
new_combine=dummies_outlet_size
new_combine=new_combine.drop(['Item_Identifier','Outlet_Identifier','Outlet_Establishment_Year','Item_Type','Item_Fat_Content'],axis=1)
new_combine
```

```python
print(new_combine.info())
```

```python
#splitting train and test data
row_train=data_train.shape[0]
new_combine.sort_values(by=['Item_Outlet_Sales'],ascending=False,inplace=True)
new_train,new_test=new_combine.iloc[:row_train,:],new_combine.iloc[row_train:,:]
new_train.tail()
new_test.drop(['Item_Outlet_Sales','Item_Count','Outlet_Count'],axis=1,inplace=True)
new_test.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
LinReg=make_pipeline(PolynomialFeatures(3),LinearRegression())
```

```python
X=new_train.drop(['Item_Outlet_Sales','Item_Count','Outlet_Count'],axis=1)
y=new_train.Item_Outlet_Sales
```

```python
LinReg.fit(X,y)
LinReg.score(X,y)
```

```python
predictiontrain=LinReg.predict(X)
predictiontrain
```

```python
from sklearn import metrics 
from sklearn.model_selection import cross_val_score
```

```python
print(metrics.mean_absolute_error(y,predictiontrain))
print(metrics.mean_squared_error(y,predictiontrain))
print(np.sqrt(metrics.mean_squared_error(y,predictiontrain)))
```

```python
predictiontest=LinReg.predict(new_test)
predictiontest
```

```python
#DecisionTree
from sklearn.tree import DecisionTreeRegressor
Classifier = DecisionTreeRegressor(criterion='mse',max_depth=10)
Classifier.fit(X,y)
Classifier.score(X,y)*100
```

```python
predict_val=Classifier.predict(X)
predict_val
```

```python
print(metrics.mean_absolute_error(y,predict_val))
print(metrics.mean_squared_error(y,predict_val))
print(np.sqrt(metrics.mean_squared_error(y,predict_val)))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor
RFClassifier = RandomForestRegressor(n_estimators=100, min_samples_split=25, max_depth=10) 
RFClassifier.fit(X,y)
RFClassifier.score(X,y)
```

```python
predict_valRF=RFClassifier.predict(X)
predict_valRF
```

```python
print(metrics.mean_absolute_error(y,predict_valRF))
print(metrics.mean_squared_error(y,predict_valRF))
print(np.sqrt(metrics.mean_squared_error(y,predict_valRF)))
```
